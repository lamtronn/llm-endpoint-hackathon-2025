```python
import os
# os.environ['KMP_DUPLICATE_LIB_OK']='True'
import torch
import numpy as np
import matplotlib.pyplot as plt
import SimpleITK as sitk
# import cv2
import os
```

```python

def read_data(path_to_nifti, return_numpy=True):
        """Read a NIfTI image. Return a numpy array (default) or `nibabel.nifti1.Nifti1Image` object"""
        # if return_numpy:
        #     return nib.load(str(path_to_nifti)).get_fdata()
        # return nib.load(str(path_to_nifti))
        return sitk.GetArrayFromImage(sitk.ReadImage(str(path_to_nifti)))

def standadize_nonzeros(image):
    img_nonzeros = image[image!=0]
    norm_img = (image - img_nonzeros.mean()) / img_nonzeros.std()
    return norm_img

def Crop(img, target_size=128, stride=128):
    img_list = []

    d, h, w = img.shape[:3]
    for z in range(0, d, stride):
        for y in range(0, h, stride):
            for x in range(0, w, stride):
                if z + target_size <= d and y + target_size <= h and x + target_size <= w:
                    crop = img[z:z + target_size, y:y + target_size, x:x + target_size, :]
                    img_list.append(crop)
                else:
                    crop = img[d - target_size:d, h - target_size:h, w - target_size:w, :]
                    img_list.append(crop)

    img_array = np.stack(img_list, axis=0)                
    return img_array
```
